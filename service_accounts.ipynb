```python
# To display the service accounts
import wmi

def get_services():
    c = wmi.WMI()
    services_with_logon_users = []

    for service in c.Win32_Service():
        service_name = service.Name
        start_name = service.StartName
        logon_user = start_name.split('\\')[-1] if start_name else "N/A"  

        services_with_logon_users.append({
            'Service Name': service_name,
            'Logon User': logon_user
        })

    return services_with_logon_users

def print_services(services):
    for service in services:
        print("Service Name:", service['Service Name'])
        print("Logon User:", service['Logon User'])
        print("=" * 40)

if __name__ == "__main__":
    services = get_services()
    print_services(services)
```

```python
#To display task Schedulers with log on users
import subprocess
import re

def task_schedulers():

    try:
        
        schtasks_command = 'schtasks /query /v /fo list'
        output = subprocess.check_output(schtasks_command, shell=True, text=True)

        
        tasks = re.split(r'\n\n', output)

    
        task_info = []

        for task in tasks:
            task_name_match = re.search(r'TaskName:\s(.+)', task)
            run_as_user_match = re.search(r'Run As User:\s(.+)', task)

            task_name = task_name_match.group(1).strip() if task_name_match else "Not Found"
            run_as_user = run_as_user_match.group(1).strip() if run_as_user_match else "Not Found"

            task_info.append({
                'Task Name': task_name,
                'Run as User': run_as_user
            })

        
        for i, info in enumerate(task_info):
            print(f"Task {i + 1} Information:")
            for key, value in info.items():
                print(f"{key}: {value}")
            print()

    except subprocess.CalledProcessError as e:
        print(f"Error running schtasks command: {e.output}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        
if __name__=="__main__" :
    task_schedulers()       
```

```python

#display the iis websites with log on users
import subprocess
import re


appcmd_path = r'C:\Windows\System32\inetsrv\appcmd.exe'

def get_iis_sites():
    
    process = subprocess.Popen([appcmd_path, "list", "site"], stdout=subprocess.PIPE, text=True)
    output, err = process.communicate()
    if err:
        raise Exception(err)

    sites = []
    site_info = re.findall(r'SITE "(.*?)" \(id:(\d+),', output)
    for name, id in site_info:
        sites.append({"Name": name, "Id": id})

    return sites

def main():
    sites = get_iis_sites()

    for site in sites:
        print("Site:", site["Name"])
        

if __name__ == "__main__":
    main()
```

```python
#display the Application app pool
import subprocess

def list_application_pools():
    
    try:
        # Specify the full path to appcmd
        appcmd_path = r'C:\Windows\System32\inetsrv\appcmd.exe'

        
        process = subprocess.Popen([appcmd_path, 'list', 'apppool'], stdout=subprocess.PIPE, text=True)
        output, err = process.communicate()

        if err:
            raise Exception(err)

        return output

    except Exception as e:
        return str(e)

if __name__ == "__main__":
    application_pools = list_application_pools()
    
    if isinstance(application_pools, str):
        print("IIS Application Pools:")
        print(application_pools)
```
